#기본 매개변수

- 함수 정의부의 매개변수에 기본값을 대입함

- 호출부에서 그 매개변수의 값을 전달하지 않으면, default값으로 사용됨.

```python
# 정의부
# n = 2 : 호출부에서 전달인자 없기 때문에 기본값으로 사용됨.


def print_n_time(value, n = 2):
  for i in range(n):
    print(value)

#호출부
print_n_time("안녕하세요.",10)

```
