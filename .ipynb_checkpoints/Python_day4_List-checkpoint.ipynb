```python
"""
주머니에 돈이 있으면 택시를 타고,
주머니에 돈은 없지만 카드가 있으면 택시를 타고,
돈도 없고 카드도 없으면 걸어 가라
#
포켓의 상태
pocket = money , paper , mobilephone
card = True
"""
"""

if 돈이 있거나 or 카드가 있거나 :
택시타기
else:
걸어가기
"""

### 주머니에 있는 방식 논리는 [주머니 안 = list]로 개념을 두고, 이 안에 무엇이 있는지 판단하는 방식으로 생각하자.
pocket = ['money' , 'paper' , 'mobilephone']
card = True
#if 돈이 있거나 or 카드가 있거나 :
if 'money' in pocket or card :
  print('택시타기')
else :
    print('걸어가기')
```

```python
"""
자판기 문제 : 돈을 넣고 => 으로 입력 받음
콜라, 사이다, 환타를 고를 수 있음 => input()으로 입력 받음
한 번에 1개만 고를 수 있음
고른 것에 따라 값을 차감하고 고른 것과 잔돈을 출력함
각각의 가격 콜라 : 500 , 사이다 600, 환타 700
만약 : 입력값이 700원보다 작을 경우, 잔돈이 부족합니다. 출력할 것
"""

#700 <= a       음료를 다 마실 수 있다
#600 <= a < 700 사이다, 콜라만 마실 수 있다.
#500 <= a < 600 콜라만 마실 수 있다.
# a  < 500      아무것도 마실 수 없다.

##고객 여정 - 돈을 넣는다. 콜라, 사이다, 환타를 고른다.

money = eval(input("동전을 넣어주세요"))

if money < 700 :
  print("잔돈이 부족합니다.")
else :
  beverage = input("콜라, 사이다, 환타 중 선택하세요")
  if beverage == "환타":
    print(money - 700)
  elif beverage == "사이다":
    print(money - 600)
  elif beverage == "콜라":
    print(money - 500)
```

```python
#반복문

data = [11, 22, 33, 'a', 'b', 'c']
```

```python
data
```

```python
index = 0
for a in data :
  print(f"인덱스{index}, 값{a}")
  index = index + 1
```

```python
index = 0
for a in data :
  print(f"인덱스{index}, 값{a}")
  index = index + 1
  print("반복문 안에 있음")
  print("-"*20)

print("반복문 밖에 있음")
```

```python
datas = input(", 숫자를 구분해서 입력 >> 1,2")
datas
```

```python
datas = datas.split(",")
datas
```

```python
#for 요소 하나 받는 변수 in 반복가능한 자료형 :

sum_result = 0

for dataa in datas:
  #더하기(sum)
  sum_result = sum_result + int(dataa)
  print(sum_result)
```

```python
## Enumerate
for index, dd in enumerate(data) :
  print(f"인덱스 {index}, 값 {dd}")
  print("반복문 안에 있음")
  print("-")*20
print("반복문 밖이야")
```

```python
thousands = list(range(1,1000))
one = []
ten = []
hundreds = []

for elements in thousands:
  if 0 < elements <10:
    one.append(elements)
  elif 10 <= elements <100:
    ten.append(elements)
  else:
    hundreds.append(elements)

print(one)
print(ten)
print(hundreds)
```
