# **문자열** 인덱싱
문자열은 자료 하나가 들어가 있는 것이다. 용량의 차이!

기본적으로 왼쪽에 있는 것에 뒤에 있는 것을 넣는 논리.

변수는 기본적으로 식별자


```python
data_str = "이승준"
```

```python
data_str
```

```python
data_str[0]
```

```python
data_str[1]
```

```python
data_str[2]
```

```python
# 인덱스 범위를 벗어나서 인덱싱을 하면 오류가 난다. 값을 불러오는 것은 인덱싱
data_str[3]
```

```python
data_str[0:]
```

```python
data_str[1:]
```

```python
# [:n] - n-1까지의 값을 모두 참조 ->  n-1까지 '슬라이싱 한다' 라고 표현한다.

data_str[:2]
```

```python
num3 = input("정보를 입력하세요")
```

```python
num3
```

```python
num4 = input("정보를 입력하세요")
```

```python
num3 = eval(num3)
type(num3)
```

```python
num4 = eval(num4)
```

```python
print(f"{num3} + {num4} = {num3 + num4}")
```

```python
print(f"{num3} + {num4} = {num3 + num4}")
print(f"{num3} - {num4} = {num3 - num4}")
print(f"{num3} * {num4} = {num3 * num4}")
print(f"{num3} / {num4} = {num3 / num4}")
print(f"{num3} % {num4} = {num3 % num4}")
print(f"{num3} // {num4} = {num3 // num4}")
```

```python
print("두 수의 사칙 연산 결과")
print(f"{num3} + {num4} = {num3 + num4}")
print(f"{num3} - {num4} = {num3 - num4}")
print(f"{num3} * {num4} = {num3 * num4}")
print(f"{num3} / {num4} = {num3 / num4}")
print(f"{num3} % {num4} = {num3 % num4}")
print(f"{num3} // {num4} = {num3 // num4}")
```
